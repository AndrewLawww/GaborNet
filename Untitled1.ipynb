```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alexnet_data = []
gabornet_data = []

with open('alexnet_test.txt', 'r') as f:
    for line in f:
        line = line.replace(')', '').replace("(", '').split()
        alexnet_data.append(line)
        
with open('gabornet_test_.txt', 'r') as f:
    for line in f:
        line = line.replace(')', '').replace("(", '').split()
        gabornet_data.append(line)
```

```python
alexnet_data = np.array(alexnet_data)
gabornet_data = np.array(gabornet_data)
```

```python
alexnet_data[0]
```

```python
num = 196
alex_mean = []
gabor_mean = []
for i in range(90):
    alex_mean.append(np.sum(alexnet_data[i*num:i*num+num, -2].astype(np.float))/num)
    gabor_mean.append(np.sum(gabornet_data[i*num:i*num+num, -2].astype(np.float))/num)
```

```python
alex_mean = np.array(alex_mean)
print(alex_mean[[0,4,9,14,24,44,69,88]])
gabor_mean = np.array(gabor_mean)
print(gabor_mean[[0,4,9,14,24,44,69,88]])
```

```python
min(alex_mean-gabor_mean)
```

```python
plt.figure(figsize=(10,10))
plt.plot(alex_mean, label = 'a')
plt.plot(gabor_mean, label = 'g')
plt.legend()
plt.title('alexnet|gabornet train')
plt.ylabel('top5')
plt.xlabel('epoch')
plt.grid()
plt.show()
```
